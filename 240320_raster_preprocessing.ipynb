```python
# This file creates raster_data.csv for top_k_bc approach

import numpy as np
import pandas as pd
```

```python
# load all raster data and combine it into one dataset

bioclimatic_df = pd.read_csv('../data/l1/EnvironmentalRasters/EnvironmentalRasters/Climate/Average 1981-2010/GLC24-PA-train-bioclimatic.csv')
bioclimatic_df
```

```python
lc_df = pd.read_csv('../data/l1/EnvironmentalRasters/EnvironmentalRasters/LandCover/GLC24-PA-train-landcover.csv')
lc_df
```
